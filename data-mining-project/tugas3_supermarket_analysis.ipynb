```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("supermarket_sales.csv")

df.head(10)

```

```python
df.info()
```

```python
df["Total"] = (df["Unit price"] * df["Quantity"]) + df["Tax"]

df.head()
```

```python
plt.figure(figsize=(8,5))
df.groupby("Branch")["Total"].mean().plot(kind="bar", color=["orange", "skyblue", "green"])

plt.title("Rata-rata Total Penjualan per Cabang")
plt.xlabel("Cabang")
plt.ylabel("Rata-rata Total Penjualan")
plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.show()
```

```python
max_total = df.loc[df["Total"].idxmax()]
min_total = df.loc[df["Total"].idxmin()]

print("Produk dengan Total Penjualan Tertinggi:")
print(max_total)

print("\nProduk dengan Total Penjualan Terendah:")
print(min_total)
```

```python
plt.figure(figsize=(7,7))

df["Product line"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    shadow=True
)

plt.title("Persentase Transaksi per Product Line")
plt.show()
```
